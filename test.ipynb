```python
import json
path = "/home/b/bipinp/projects/llms/SynthPAI/data/synthpai_merged_evals.jsonl"
json_list = []
json_list = []
with open(path, "r") as json_file:
    for line in json_file:
        json_obj = json.loads(line)
        json_list.append(json_obj)
        
```
